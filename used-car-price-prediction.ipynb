# Label Encoding
### Our dataset has 12 characteristic categorical variables and 4 numerical variables (not including the price column). To apply ML models, we need to convert these categorical variables into numerical variables and LabelEncoder software from sklearn library is used to solve this problem. The below figure depicts label encoding which is performed on the dataset.

```python
#Label Encoding is used for the following columns.

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
X=['manufacturer','fuel','transmission','type','drive','condition',
   'cylinders','title_status'
  ,'state','paint_color','size']
   #,'state','paint_color','size']
for x in X:
    dataframe[x] = le.fit_transform(dataframe[x])
```

```python
dataframe
```

# Imputation
### The Iterative Imputer method is used to impute missing values and different estimators are applied, and the MSE of each estimator is calculated using cross_val_score.

```python
#imputation is done in order to remove null values in the dataset. This will help us in acheiving a good accuracy for the ML Model.
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
ds=['cylinders','price','manufacturer','condition','fuel','title_status','year','transmission','drive','type','odometer'
    ,'state','paint_color','size']

imputer = IterativeImputer(imputation_order='ascending',random_state=1)
imputer.fit(dataframe[ds])
dataframe[ds]=imputer.transform(dataframe[ds])
```

```python
#Removing outliers with the help of boxplot

na='price'
pIQR=dataframe[na].quantile(0.75)-dataframe[na].quantile(0.25)
pupper=dataframe[na].quantile(0.75)+1.5*pIQR
plower=dataframe[na].quantile(0.25)-1.5*pIQR

print(na)
print(dataframe[na].quantile(0.75))
print(dataframe[na].quantile(0.05))
print(pupper)
print(plower)


na='odometer'
oIQR=dataframe[na].quantile(0.75)-dataframe[na].quantile(0.25)
oupper=dataframe[na].quantile(0.75)+1.5*oIQR
olower=dataframe[na].quantile(0.25)-1.5*oIQR

print(na)
print(dataframe[na].quantile(0.75))
print(dataframe[na].quantile(0.05))
print(oupper)
print(olower)

na='year'
yIQR=dataframe[na].quantile(0.75)-dataframe[na].quantile(0.25)
yupper=dataframe[na].quantile(0.75)+1.5*yIQR
ylower=dataframe[na].quantile(0.25)-1.5*yIQR

print(na)
print(dataframe[na].quantile(0.75))
print(dataframe[na].quantile(0.05))
print(yupper)
print(ylower)
```

```python
dataframe=dataframe[dataframe.price<pupper]
dataframe=dataframe[dataframe.price>1000]


dataframe=dataframe[dataframe.odometer<oupper]

dataframe=dataframe[dataframe.odometer>olower]

dataframe=dataframe[dataframe.year<yupper]

dataframe=dataframe[dataframe.year>ylower]
```

```python
#Dropping the target column in the dataset
y = dataframe['price']
x = dataframe.drop('price',axis= 1)
```

```python
#Splitting the dataset into training (80%)and testing data (20%) for model building phase
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error,median_absolute_error
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=10)
```

# Regression Model and Prediction

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import warnings
warnings.filterwarnings('ignore')

output=[]
output.append(["Regression Algorithm","RMSE","R Squared","Mean Squared Error","Absolute Mean Error","Median e"])
dictout={}
dictout1={}
dictout2={}
```

```python
#Each of the Machine learning models used will give us mean square error, mean absoulute error, R-squared error, root mean squared error and Median ame with the help of the below function

def train_test_model(model,model_name,X_train,y_train,X_test,y_test):
    model.fit(X_train,y_train)
    predictions = model.predict(X_test)
    
    plt.figure(figsize=(10,10))
    plt.xlabel("Actual price")
    plt.ylabel("Predicted price")
    plt.title(model_name)
    plt.plot( y_test, predictions,'b.',alpha=0.05)
    plt.plot(y_test,y_test,'r-')
    plt.legend(['Predicted pric','Actual Price'],loc='upper left')
    plt.show()
    '''
    a=[]
    j=0
    for i in y_test.index:
        if abs(predictions[j]-y_test[i])>30000:
            a.append(i)
        j=j+1
    t=X_test.loc[a]
    #t.describe()
    print( t.describe())
    for i in t.columns:
        print(t[i].value_counts())
        
    '''
    mse=mean_squared_error(y_test, predictions)
    mae=mean_absolute_error(y_test, predictions)
    rs=model.score(X_test,y_test, sample_weight=None)
    rmse=math.sqrt(mse)
    maelog=median_absolute_error(y_test,predictions)
    print(model_name,'\nMean Squared error : ', mse)
    print('Mean Absolute error : ', mae)
    print("R-Squared error",rs)
    print("Root Mean Squared Error: ",rmse)
    print("Median ame",maelog)

    output.append([model_name,rmse,rs,mse,mae,maelog])
    dictout[model_name]=rmse
    dictout1[model_name]=rs
    dictout2[model_name]=maelog
```

## Multiple Linear regression
### Multiple Linear regression model is the quickest and easiest regression model but the precision is lower than other models and used only in certain problems. It uses weights to fit the training data in a straight-line and weights are adjusted by minimizing the mean squared error loss.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
train_test_model(model,"Linear Regression",X_train,y_train,X_test,y_test)
```

## Ridge Regression
### Ridge regression is a regularized version of linear regressor but has alpha parameter to control the regulation of the model to reduce the variance of the estimates.

```python
from sklearn.linear_model import Ridge
model = Ridge(alpha=1,fit_intercept=True,random_state=1)
train_test_model(model,"Ridge Regression",X_train,y_train,X_test,y_test)
```

## LASSO Regression
### LASSO stands for Least Absolute Shrinkage and Selection Operator which is a shrinkage regularization technique. It is a simple sparse model which has only fewer parameters than ridge or linear regression. This type of regression is used when the dataset shows high multi collinearity or when you want to automate variable elimination and feature selection.


```python
from sklearn.linear_model import Lasso

model = Lasso(fit_intercept=True,random_state=1)
train_test_model(model,"Lasso Regression",X_train,y_train,X_test,y_test)
```

## Bayesian Regression
### Bayesian regression is a type of linear regression which can be implemented with insufficient data or poorly distributed data using probability distributors rather than point estimates.

```python
from sklearn.linear_model import BayesianRidge
model=BayesianRidge(fit_intercept=True)

train_test_model(model,"Bayesian Regression",X_train,y_train,X_test,y_test)
```

## Decision Tree
## Decision Tree a widely used model to handle missing values with high dimensional data. It is more effective and provides optimum accuracy with minimum number of features.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
model = DecisionTreeRegressor(max_depth=23,max_features='auto',random_state=1)

train_test_model(model,"Decision Tree Regression",X_train,y_train,X_test,y_test)
```

## Random Forest Regression
### Random forest regression is a classifier based on ensemble Machine Learning model which is used in decision tree framework to create a random decision tree. It has a better prediction accuracy compared to normal regression.

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100,max_depth=33,max_features='log2',random_state=1)
train_test_model(model,"Random Forest Regression",X_train,y_train,X_test,y_test)
```

## Multi-layer Perceptron 
### Multi-layer Perceptron (MLP) Regression Algorithm is a type of supervised learning which emulate a function from the input dataset. It is an artificial neural network composed of multiple layers of perceptron. It uses hidden layers and neurons to predict the value of the unknown input.

```python
from sklearn.neural_network import MLPRegressor
model=MLPRegressor(hidden_layer_sizes=(28,28,42,1),activation='relu',learning_rate='adaptive', learning_rate_init=0.1, solver='adam', max_iter=320,random_state=1,verbose=True,early_stopping=True,n_iter_no_change=70)
train_test_model(model,"MLP Regression",X_train,y_train,X_test,y_test)
```

## Light gradient-boosting
### Light gradient-boosting ensemble machine learning models which has XG Boost advantages.

```python
from lightgbm import LGBMRegressor
model=LGBMRegressor(max_depth=16,random_state=1)

train_test_model(model,"LGBM Regression",X_train,y_train,X_test,y_test)
```

## XG Boost
### XG Boost ensemble learning models uses loss function and a regularization term.

```python
import xgboost as xg
model=xg.XGBRegressor(max_depth=14,random_state=1)

train_test_model(model,"XGB Regression",X_train,y_train,X_test,y_test)
```

## Gradient Boosted Regression
### Gradient boosted regression is ensemble model same as the random forest regression but in gradient boosted it creates multiple trees to improve the deficiencies in the previous tree so absolute mean error will be the least. It can be modified with the learning rate and loss function. It is important to note that the learning rate should not be high so the tree will not be overfitted.

```python
from sklearn.ensemble import GradientBoostingRegressor
model=GradientBoostingRegressor(max_depth=16,random_state=1)

train_test_model(model,"Gradient Boost Regression",X_train,y_train,X_test,y_test)
```
