# Label Encoding

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
X=['manufacturer','fuel','transmission','type','drive','condition',
   'cylinders','title_status'
  ,'state','paint_color','size']
   #,'state','paint_color','size']
for x in X:
    dataframe[x] = le.fit_transform(dataframe[x])
```

# Data Splitting

```python
dataframe
```

```python
y = dataframe['price']
x = dataframe.drop('price',axis= 1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error,median_absolute_error
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3,random_state=1)
```

# Regression Model and Prediction

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import warnings
warnings.filterwarnings('ignore')

output=[]
output.append(["Regression Algorithm","RMSE","R Squared","Mean Squared Error","Absolute Mean Error","Median e"])
dictout={}
dictout1={}
dictout2={}
```

```python
def train_test_model(model,model_name,X_train,y_train,X_test,y_test):
    model.fit(X_train,y_train)
    predictions = model.predict(X_test)
    
    plt.figure(figsize=(10,10))
    plt.xlabel("Actual price")
    plt.ylabel("Predicted price")
    plt.title(model_name)
    plt.plot( y_test, predictions,'b.',alpha=0.05)
    plt.plot(y_test,y_test,'r-')
    plt.legend(['Predicted pric','Actual Price'],loc='upper left')
    plt.show()
    '''
    a=[]
    j=0
    for i in y_test.index:
        if abs(predictions[j]-y_test[i])>30000:
            a.append(i)
        j=j+1
    t=X_test.loc[a]
    #t.describe()
    print( t.describe())
    for i in t.columns:
        print(t[i].value_counts())
        
    '''
    mse=mean_squared_error(y_test, predictions)
    mae=mean_absolute_error(y_test, predictions)
    rs=model.score(X_test,y_test, sample_weight=None)
    rmse=math.sqrt(mse)
    maelog=median_absolute_error(y_test,predictions)
    print(model_name,'\nMean Squared error : ', mse)
    print('Mean Absolute error : ', mae)
    print("R-Squared error",rs)
    print("Root Mean Squared Error: ",rmse)
    print("Median ame",maelog)

    output.append([model_name,rmse,rs,mse,mae,maelog])
    dictout[model_name]=rmse
    dictout1[model_name]=rs
    dictout2[model_name]=maelog
```

## Multiple Linear regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
train_test_model(model,"Linear Regression",X_train,y_train,X_test,y_test)
```

## Ridge Regression


```python
from sklearn.linear_model import Ridge
model = Ridge(alpha=1,fit_intercept=True,random_state=1)
train_test_model(model,"Ridge Regression",X_train,y_train,X_test,y_test)
```

## LASSO Regression

```python
from sklearn.linear_model import Lasso

model = Lasso(fit_intercept=True,random_state=1)
train_test_model(model,"Lasso Regression",X_train,y_train,X_test,y_test)
```

## Bayesian Regression

```python
from sklearn.linear_model import BayesianRidge
model=BayesianRidge(fit_intercept=True)

train_test_model(model,"Bayesian Regression",X_train,y_train,X_test,y_test)
```

## Decision Tree

```python
print('Decision Tree Regressor with different depth')
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree

for k in range(15,20):
    model = DecisionTreeRegressor(max_depth=k,max_features='auto',random_state=1)
    model.fit(X_train,y_train)
    predictions = model.predict(X_test)
    print('Depth',k,'RMSE : ',  math.sqrt(mean_squared_error(y_test, predictions)),'   R SQUARE:',model.score(X_test,y_test, sample_weight=None))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
model = DecisionTreeRegressor(max_depth=18,max_features='auto',random_state=1)

train_test_model(model,"Decision Tree Regression",X_train,y_train,X_test,y_test)
```

## Multi-layer Perceptron

```python
from sklearn.neural_network import MLPRegressor
model=MLPRegressor(hidden_layer_sizes=(28,28,42,1),activation='relu',learning_rate='adaptive', learning_rate_init=0.1, solver='adam', max_iter=320,random_state=1,verbose=True,early_stopping=True,n_iter_no_change=70)
train_test_model(model,"MLP Regression",X_train,y_train,X_test,y_test)
```
